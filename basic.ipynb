OOP = Organizing code like real-world objects.

Instead of writing random functions, you model entities with data + behavior.

                                1️⃣ Class and Object
                                🔹 Logic

                                Class → Blueprint

                                Object → Real instance of blueprint

                                Example:
                                Class = Car design
                                Object = Your actual car

```python
class Student:
    def __init__(self, name, mark):
        self.name = name
        self.mark = mark

    def display(self):
        print("Name:", self.name)
        print("Mark:", self.mark)

s1 = Student("Nanba", 85)
s1.display()
```

Constructor

Parameterized constructor

```python
class Car:
    def __init__(self, brand="Unknown"):
        self.brand = brand

c1 = Car()
c2 = Car("BMW")

print(c1.brand)
print(c2.brand)
```

Instance Variables vs Class Variables
| Type              | Belongs To | Memory        |
| ----------------- | ---------- | ------------- |
| Instance variable | Object     | Separate copy |
| Class variable    | Class      | Shared        |


```python
class Employee:
    company = "TCS"   # Class variable

    def __init__(self, name):
        self.name = name  # Instance variable

e1 = Employee("Ram")
e2 = Employee("Shyam")

print(e1.company)
print(e2.company)
```

Methods in Class
🔹 Types

Instance method

Class method

Static method

```python
class Demo:
    count = 0

    def __init__(self):
        Demo.count += 1

    def instance_method(self):
        print("Instance Method")

    @classmethod
    def class_method(cls):
        print("Total objects:", cls.count)

    @staticmethod
    def static_method():
        print("Static Method")

d1 = Demo()
d2 = Demo()

Demo.class_method()
```

                            5️⃣ Encapsulation
                                    🔹 Logic

                                    Data hiding. Protect internal data.

                                    Python doesn’t truly make variables private, but naming convention helps.

```python
class Bank:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

b = Bank(1000)
b.deposit(500)
print(b.get_balance())
```

```python
class Account:
    bank_name = "Indian National Bank"   # Class Variable

    def __init__(self, acc_no, name, balance):
        self.acc_no = acc_no
        self.name = name
        self.__balance = balance   # Encapsulation

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Deposit Successful")
        else:
            print("Invalid Amount")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawal Successful")
        else:
            print("Insufficient Balance")

    def get_balance(self):
        return self.__balance

    def display(self):
        print(f"Account No: {self.acc_no}")
        print(f"Name: {self.name}")
        print(f"Balance: {self.__balance}")
class SavingsAccount(Account):
    interest_rate = 0.05

    def add_interest(self):
        interest = self.get_balance() * SavingsAccount.interest_rate
        self.deposit(interest)

    def display(self):
        print("---- Savings Account ----")
        super().display()
class CurrentAccount(Account):
    overdraft_limit = 5000

    def withdraw(self, amount):
        if amount <= self.get_balance() + CurrentAccount.overdraft_limit:
            print("Withdrawal Successful (Overdraft Allowed)")
        else:
            print("Overdraft Limit Exceeded")

    def display(self):
        print("---- Current Account ----")
        super().display()
class Bank:
    def __init__(self):
        self.accounts = []

    def create_account(self, account):
        self.accounts.append(account)
        print("Account Created Successfully")

    def show_all_accounts(self):
        for acc in self.accounts:
            acc.display()
            print("-------------------")
bank = Bank()

acc1 = SavingsAccount(101, "Nanba", 10000)
acc2 = CurrentAccount(102, "Arun", 5000)

bank.create_account(acc1)
bank.create_account(acc2)

acc1.deposit(2000)                                                                                                                                                                                                                                                                                                                                                                                                  
acc1.add_interest()
                                                                        
acc2.withdraw(7000)

bank.show_all_accounts()
```
