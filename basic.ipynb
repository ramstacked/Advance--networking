                🔥 Python OOP – Full Concepts with Logic + Programs

OOP = Organizing code like real-world objects.

Instead of writing random functions, you model entities with data + behavior.

                                1️⃣ Class and Object
                                🔹 Logic

                                Class → Blueprint

                                Object → Real instance of blueprint

                                Example:
                                Class = Car design
                                Object = Your actual car

```python
class Student:
    def __init__(self, name, mark):
        self.name = name
        self.mark = mark

    def display(self):
        print("Name:", self.name)
        print("Mark:", self.mark)

s1 = Student("Nanba", 85)
s1.display()
```
